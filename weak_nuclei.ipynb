```python
import pynucastro as pyna
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica"
})

plt.rcParams['figure.figsize'] = [4, 4]

plt.rcParams['savefig.bbox']='tight'
```

```python
# plt.xticks(fontsize=36)
# plt.yticks(fontsize=36)

lib = pyna.SuzukiLibrary()
net = pyna.RateCollection(libraries=[lib])

net.gridplot(cmap="coolwarm", minN=0 , maxN=40, minZ=0, maxZ=40, dpi=600)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig("pyna_20.pdf")
```

```python
lib2 = pyna.TabularLibrary()
net2 = pyna.RateCollection(libraries=[lib2])

net2.gridplot(cmap="PuOr", minN=0, maxN=40, minZ=0, maxZ=40, dpi=600)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig("pyna_21.pdf")
```
