```python
import pynucastro as pyna
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica"
})

plt.rcParams['savefig.bbox']='tight'
```

```python
lib = pyna.SuzukiLibrary()
net = pyna.RateCollection(libraries=[lib])
net.gridplot()

#plt.title(r"{\tt pynucastro} $2.1$ available weak rate nuclei")
plt.xlim(0,40)
plt.ylim(0,40)
```
